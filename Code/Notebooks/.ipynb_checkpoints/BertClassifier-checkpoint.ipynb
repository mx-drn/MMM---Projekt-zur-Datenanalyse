```python
#Dictionary mit allen .json
FILES = {1: 'Gangart.json',
        2 : 'Vegan.json',
        3 : 'Laender.json',
        4 : 'Mahlzeit.json'}

MODELS = {1:'wwm_uncased_L-24_H-1024_A-16',   #bert-large
          2:'multi_cased_L-12_H-768_A-12'}    #bert-base
```

```python
#Globale Paramater festlegen

#Ausgewählter Datensatz
CURRENT_FILE = FILES[2]

#Ausgewähltes Bert-Model
MODEL_FILE = MODELS[2]

#Aktuelles Verzeichnis
PATH = 'drive/MyDrive/Colab Notebooks/'

#Zutaten als String vor die Anweisung hängen
BERT_WITH_INGREDIENTS = True

#Nur Bert oder nur MLP oder beide verwenden
BOTH_NEURAL_NETWORKS = True

#Minimale Länge aller Directions
MINLEN_DIRECTIONS = 106

#Hyperparamter festlegen
DROPOUT_RATE = 0.5
EPOCHS = 8
BATCH_SIZE = 64
LOSS_FUNCTION = 'sparse_categorical_crossentropy'
OPTIMIZER = tf.optimizers.Adam(lr=0.00001)

#Menge Units der einzelnen Neuronalen Netze
OUTPUT_SHAPE_ZUTATEN = 32
OUTPUT_SHAPE_ANWEISUNGEN = 32
```

```python
DATA_PATH = '../Data/'
#Load Recipes from .json file
with open(DATA_PATH + CURRENT_FILE) as f:
    data = json.load(f)
```

```python
#Directions aufbereiten
def preprocess_text(sen):

    # Punkte und Zahlen eliminieren
    sen = re.sub('[^a-zA-Z]', ' ', sen)

    # Einzelne Characters entfernen
    sen = re.sub(r"\s+[a-zA-Z]\s+", ' ', sen)

    # Mehrere Leerzeichen entfernen
    sen = re.sub(r'\s+', ' ', sen)

    #Stopwords entfernen
    sen = ' '.join(filter(lambda word: word.lower() not in stopwords.words('english'),  sen.split()))

    return sen

TAG_RE = re.compile(r'<[^>]+>')
```

```python
#Input und Output Vektor
inputData = []
labels = []

#String für WordCloud
inputWordCloud = []

#Dictionaries
index_Kategorien = dict()
index_Zutaten = dict()

#Create Dict for Ingredients and Categories and Directions with index as value
def createDict():
    counterIng = 0
    counterCat = 0
    
    for recipe in data:
        ingredients = recipe['ingredients']
        category = recipe['category']
        anweisungen = recipe['directions_string']

        #index_Zutaten befüllen
        for ing in ingredients:
            if ing not in index_Zutaten:
                index_Zutaten[ing] = counterIng
                counterIng +=1
        
        #Index_Kategorien befüllen
        if category not in index_Kategorien:
            index_Kategorien[category] = counterCat
            counterCat +=1 
          
 
#Create Input Vector and Output Vector (OneHot)
def createDataAndLabel():
    for recipe in data:
        
        #Input Vector
        ingredients = recipe['ingredients']
        category = recipe['category']
        anweisungen = recipe['directions_string']
        
        inputVectorZutaten = np.zeros(len(index_Zutaten))
        inputZutatenString = ""
        
        #InputVector Zutaten erstellen
        for zutat in ingredients:
            if zutat in index_Zutaten:
                inputVectorZutaten[index_Zutaten[zutat]] = 1
                inputZutatenString += " " + zutat
                inputWordCloud.append(zutat)
        
        if BERT_WITH_INGREDIENTS:
          inputData.append([inputVectorZutaten, preprocess_text(inputZutatenString) + preprocess_text(anweisungen)])
        else:
          inputData.append([inputVectorZutaten,preprocess_text(anweisungen)])
        
        #Output Vector
        output = index_Kategorien[category]
        labels.append(output)
        
createDict()
createDataAndLabel()  
```

```python
def splitData(trainSize=0.8,testsize=0.2):
  #Daten aufsplitten
  train_input, test_input, train_label, test_label = train_test_split(inputData, labels, test_size=0.2, random_state=42)

  #Input Vektor für Zutaten und Anweisungen trennen
  train_input_zutaten = []
  train_input_anweisungen = []
  test_input_zutaten = []
  test_input_anweisungen = []

  for entry in train_input:
      train_input_zutaten.append(entry[0])
      train_input_anweisungen.append(entry[1])
      
  for entry in test_input:
      test_input_zutaten.append(entry[0])
      test_input_anweisungen.append(entry[1])

  #Als numpy-array
  train_label = np.array(train_label)
  test_label = np.array(test_label)
  train_input_zutaten = np.array(train_input_zutaten)
  test_input_zutaten = np.array(test_input_zutaten)

  return train_label, test_label, train_input_zutaten, test_input_zutaten,train_input_anweisungen, test_input_anweisungen

train_label, test_label, train_input_zutaten, test_input_zutaten,train_input_anweisungen,test_input_anweisungen = splitData()

#Wichtigkeit der Kategorien im Trainingsprozess
#sample_weight = compute_sample_weight(class_weight='balanced', y=train_label)
```

```python
#BertTokenizer laden
def createTokenizer():
    vocab_file = os.path.join(DATA_PATH + 'BERT/', MODEL_FILE + "/vocab.txt")
    tokenizer = bert.bert_tokenization.FullTokenizer(vocab_file, do_lower_case=True)
    return tokenizer

bert_tokenizer = createTokenizer()
```

```python
#Maximale Länge der Directions bestimmen
def findMaxLength(lst): 
    maxLen = -1
    for liste in lst:
        cur = len(liste)
        if cur > maxLen:
            maxLen = cur
    return maxLen
```

```python
#Weitere Parameter festlegen
MAXLEN = min(MINLEN_DIRECTIONS,max(findMaxLength(train_input_anweisungen),findMaxLength(test_input_anweisungen)))
INPUT_SHAPE_ANWEISUNGEN = MAXLEN
INPUT_SHAPE_ZUTATEN = len(index_Zutaten)
OUTPUT_SHAPE = len(index_Kategorien)
```

```python
#Daten tokenizen
def prepareData(tokenizer,train_input_anweisungen,test_input_anweisungen):
    train_tokens = [["[CLS]"] + tokenizer.tokenize(sentence) + ["[SEP]"] for sentence in train_input_anweisungen]
    train_tokens_ids = [tokenizer.convert_tokens_to_ids(token) for token in train_tokens]
    train_tokens_ids = pad_sequences(train_tokens_ids, maxlen=MAXLEN, dtype="long", truncating="post", padding="post")
    
    test_tokens = [["[CLS]"] + tokenizer.tokenize(sentence) + ["[SEP]"] for sentence in test_input_anweisungen]
    test_tokens_ids = [tokenizer.convert_tokens_to_ids(token) for token in test_tokens]
    test_tokens_ids = pad_sequences(test_tokens_ids, maxlen=MAXLEN, dtype="long", truncating="post", padding="post")

    return train_tokens_ids,test_tokens_ids

train_input_anweisungen, test_input_anweisungen = prepareData(bert_tokenizer,train_input_anweisungen,test_input_anweisungen)
```

```python
#BertLayer laden underzeugen
def createBertLayer(max_seq_length):
    global bert_layer

    bertDir = os.path.join(PATH,MODEL_FILE)
    bert_params = bert.params_from_pretrained_ckpt(bertDir)
    bert_layer = bert.BertModelLayer.from_params(bert_params, name="bert")
      
    model_layer = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(max_seq_length,), dtype='int32', name='input_ids'),
        bert_layer
    ])

    model_layer.build(input_shape=(None, max_seq_length))
    bert_layer.apply_adapter_freeze()

def loadBertCheckpoint():
    checkpointName = os.path.join(DATA_PATH + 'BERT/', MODEL_FILE + "/bert_model.ckpt")
    bert.load_stock_weights(bert_layer, checkpointName)
```

```python
def createLayer():
    createBertLayer(MAXLEN)
    loadBertCheckpoint()


createLayer()
```

```python
#Methode zur Erstellung des Zutaten Neuronalen Netzes (MLP)
def createZutatenNN(input_shape,p_dropout,output_shape):
    model = Sequential()
    model.add(Dense(128, input_dim=input_shape, activation='relu'))
    model.add(Dropout(p_dropout))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(p_dropout))
    model.add(Dense(output_shape, activation='softmax'))
    return model
```

```python
#Methode zur Erstellung des Anweisungen Convolutional Neuronalen Netzes (Bert)
def createAnweisungenNN(input_shape,p_dropout,output_shape):
    activation = ""
    if BOTH_NEURAL_NETWORKS:
        activation = 'relu'
    else:
        activation = 'softmax'
    
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(input_shape,), dtype='int32', name='input_ids'),
        bert_layer,
        tf.keras.layers.Lambda(lambda x: x[:, 0, :]),
        tf.keras.layers.Dense(output_shape, activation=activation)
    ])
    model.build(input_shape=(None, input_shape))
    return model
```

```python
#Neuronalen Netze erstellen und kompilieren
def createModel(summary=False):
    #Beide NN verwenden
    if BOTH_NEURAL_NETWORKS:
      zutatenNN = createZutatenNN(input_shape=INPUT_SHAPE_ZUTATEN,p_dropout=DROPOUT_RATE,output_shape=OUTPUT_SHAPE_ZUTATEN)
      anweisungenNN = createAnweisungenNN(input_shape=INPUT_SHAPE_ANWEISUNGEN,p_dropout=DROPOUT_RATE,
                                      output_shape=OUTPUT_SHAPE_ANWEISUNGEN)
    
      #Kombiniertes Neuronales Netz erstellen
      combinedInput = concatenate([zutatenNN.output, anweisungenNN.output])
      x = Dense(32, activation="relu")(combinedInput)
      x = Dense(OUTPUT_SHAPE, activation="softmax")(x)
      model = Model(inputs=[zutatenNN.input, anweisungenNN.input], outputs=x)
    
    #Nur Bert verwenden
    else:
      model = createAnweisungenNN(input_shape=INPUT_SHAPE_ANWEISUNGEN,p_dropout=DROPOUT_RATE,output_shape=OUTPUT_SHAPE)
      
    model.compile(loss=LOSS_FUNCTION, optimizer=OPTIMIZER,metrics=['accuracy'])
  
  if summary:
    model.summary()

  return model
```

```python
model = createModel(True)
```

```python
#Model trainieren
print('Trainieren auf ' + str(len(train_input_anweisungen)) + " Datensätzen.")
print('Testen auf ' + str(len(test_input_anweisungen)) + " Datensätzen. \n" )

if BOTH_NEURAL_NETWORKS:
  history = model.fit(x=[train_input_zutaten,train_input_anweisungen], y=train_label,
          validation_data=([test_input_zutaten,test_input_anweisungen], test_label),
          epochs=EPOCHS, batch_size=BATCH_SIZE)#,sample_weight=sample_weight)
else:
  history = model.fit(train_input_anweisungen,train_label,epochs=EPOCHS,batch_size=BATCH_SIZE,validation_data=(test_input_anweisungen, test_label))#,,sample_weight=sample_weight)
```

```python
accuracy,val_accuracy,f1Score = [],[],[]

def appendScores(history):
  if USE_MLP and USE_BERT:
    y_pred = model.predict([test_input_zutaten,test_input_anweisungen])
  elif USE_MLP:
    y_pred = model.predict(test_input_zutaten)
  else:
    y_pred = model.predict(test_input_anweisungen)
    
  y_pred = [innerList.argmax() for innerList in y_pred]

  f1Score.append(f1_score(test_label,y_pred,average='macro'))
  accuracy.append(history.history['accuracy'][-1])
  val_accuracy.append(history.history['val_accuracy'][-1])

appendScores(history)
```

```python
#Confusion Matrix erzeugen
def plot_confusion_matrix(cm, classes,normalize=True,title='Confusion matrix',cmap=plt.cm.Blues):

    #Prozent anstatt absolute Anzahl
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    
    plt.figure(figsize=(10,10))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title,fontsize=15)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.yticks(tick_marks, classes)
    classes.reverse()
    plt.xticks(tick_marks,classes, rotation=45)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label',fontsize=13)
    plt.xlabel('Predicted label',fontsize=13)
    plt.grid(False)
    plt.tight_layout()


#Predictions für Testdaten erzeugen und höchste Wahrscheinichkeit raussuchen
if BOTH_NEURAL_NETWORKS:
  y_pred = model.predict([test_input_zutaten,test_input_anweisungen])
else:
  y_pred = model.predict(test_input_anweisungen)

y_pred = [innerList.argmax() for innerList in y_pred]
cm = confusion_matrix(test_label, y_pred)
cm = np.array([np.flip(inner) for inner in cm])
plot_confusion_matrix(cm=cm, classes=list(index_Kategorien.keys()),title='Confusion Matrix unseres Klassifizieres')
```
